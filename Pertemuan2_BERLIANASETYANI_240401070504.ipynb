```python
import pandas as pd
import numpy as np


# Muat Dataset Titanic
URL = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(URL)

# Eksplorasi Awal
print(df.shape, df.columns.tolist())
print(df.isnull().sum())


# Filtering - Penumpang Wanita Kelas 1
wk1 = df[(df['Sex'] == 'female') & (df['Pclass'] == 1)]
print(wk1['Survived'].mean(), wk1['Age'].mean())

# Groupby - Tingkat kesalamatan per kelas
hasil = df.groupby('Pclass').agg(
    total=('PassengerId', 'count'),
    selamat=('Survived', 'mean')
).round(3)
print(hasil)
```

```python
import pandas as pd
import numpy as np


# Muat Dataset Titanic
URL = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(URL)

# Q1: Berapa jumlah baris dan kolom pada dataset Titanic? Kolom apa saja yang tersedia?
print(f"Q1")
print(f"Jumlah baris: {df.shape[0]}")
print(f"Jumlah kolom: {df.shape[1]}")
print(f"Nama kolom: {', '.join(df.columns)}")

# Q2: Kolom mana saja yang memiliki missing values? Berapa jumlah dan proporsinya (%) per kolom?
missing_count = df.isnull().sum()
missing_pct   = df.isnull().mean() * 100
missing_df = pd.DataFrame({
    'Jumlah Missing': missing_count,
    'Proporsi (%)': missing_pct.round(2)
})
print("=" * 55)
print("Q2")
print(missing_df[missing_df['Jumlah Missing'] > 0])

# Q3: Berapa persen penumpang yang selamat (Survived = 1)?
print("=" * 55)
print("Q3")
pct = (df['Survived'] == 1).mean() * 100
print(f"Penumpang yang selamat {pct}%")

# Q4: Filter penumpang wanita dari kelas 1 (Pclass=1). Berapa jumlahnya dan berapa rata-rata usianya?
print("=" * 55)
print("Q4")
wk1 = df[(df['Sex'] == 'female') & (df['Pclass'] == 1)]
avg_age = wk1['Age'].mean()
print(f"Jumlah penumpang wanita dari kelas 1: {len(wk1)}")
print(f"Rata-rata usia wanita dari kelas 1: {avg_age}")

# Q5: Gunakan groupby untuk menghitung tingkat keselamatan (mean Survived) per kelas penumpang. Kelas mana yang paling tinggi?
print("=" * 55)
print("Q5")
survival_rate = df.groupby('Pclass')['Survived'].mean().round(4)
print(survival_rate.to_string())
best_class = survival_rate.idxmax()
print(f"Kelas paling tinggi keselamatannya: Kelas {best_class} "
      f"({survival_rate[best_class]}%)")

#Q6: Analisis tambahan
print("=" * 55)
print("Q6")
print("Tingkat keselamatan berdasarkan jenis kelamin:")
surv_sex = df.groupby('Sex')['Survived'].mean().mul(100).round(2)
print(surv_sex.to_string())
print("Wanita memiliki peluang selamat jauh lebih tinggi dari pria")
```
